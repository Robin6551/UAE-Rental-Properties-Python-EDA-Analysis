```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('dubai_properties.csv')
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df['Posted_date'] = pd.to_datetime(df['Posted_date'])
```

```python
df[['Rent','Baths','Beds','Area_in_sqft']].describe()
```

```python
df.isnull().sum()
```

```python
df['Latitude'] = df['Latitude'].ffill()
```

```python
df['Longitude'] = df['Longitude'].ffill()
```

```python
df.duplicated().sum()
```

```python
df.boxplot(column= 'Rent')
plt.title('Rent')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
df['Type'].value_counts().plot(kind= 'bar')
plt.title('Distribution Of Type')
plt.xlabel('Type')
plt.ylabel('Total')
plt.show()
```

```python
df['Furnishing'].value_counts().plot(kind= 'pie', autopct= '%1.1f%%')
plt.title('Distribution of Furnishing')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
df['Rent_category'].value_counts().plot(kind= 'bar', color= 'green')
plt.title('Distribution Of Type')
plt.xlabel('Rent_category')
plt.ylabel('Total')
plt.show()
```

```python
df.columns
```

```python
df.groupby('Beds')['Rent'].sum().sort_values(ascending= False).plot(kind= 'bar', color= 'green')
plt.title('Relationship between Beds and Rent')
plt.xlabel('Beds')
plt.ylabel('Rent')
plt.show()
```

```python
Area_in_sqft_by_rent = df.groupby('Area_in_sqft')['Rent'].sum().sort_values(ascending=False)

# Converting the Series to a DataFrame for easier plotting.
df_plot = Area_in_sqft_by_rent.reset_index()

# Creating the scatter plot
plt.figure(figsize=(5, 4))
plt.scatter(df_plot['Area_in_sqft'], df_plot['Rent'])
plt.title('Relationship Between Area and Total Rent')
plt.xlabel('Area in Square Feet')
plt.ylabel('Total Rent')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df.groupby('Furnishing')['Rent'].sum().sort_values(ascending= False).plot(kind= 'pie',autopct= '%1.1f%%')
plt.title('Relationship between Furnishing and Rent')
plt.xlabel('Furnishing')
plt.ylabel('Rent')
plt.show()
```

```python
df.groupby('Type')['Rent'].sum().sort_values(ascending= False).plot(kind= 'bar',color= 'green')
plt.title('Type VS Rent')
plt.xlabel('Type')
plt.ylabel('Rent')
plt.show()
```

```python
df.groupby('City')['Rent'].mean().sort_values(ascending= False).plot(kind= 'bar', color= 'green')
plt.title('Relationship between Beds and Rent')
plt.xlabel('City')
plt.ylabel('Rent')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5,3))
scatter = plt.scatter(
    df['Longitude'], 
    df['Latitude'], 
    c=df['Rent'], 
    cmap='viridis', 
    alpha=0.6
)

plt.colorbar(scatter, label="Rent")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Geographical Rent Distribution in UAE")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.boxplot(x="City", y="Rent_per_sqft", data=df)
plt.title("Rent per sqft distribution by City")
plt.xticks(rotation=45)
plt.show()
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Compute correlation matrix
corr = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Numeric Variables")
plt.show()
```
